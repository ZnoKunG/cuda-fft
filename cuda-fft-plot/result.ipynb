```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
```

```python
import os

dft_n = []
dft_cpu_rows = []
dft_gpu_rows = []

with open('../cuda-fft/result_dft.csv') as result_dft:
    reader = csv.reader(result_dft, quotechar='|')
    for row in reader:
        dft_n.append(row[0])
        dft_cpu_rows.append(float(row[1]))
        dft_gpu_rows.append(float(row[2]))
        
fft_n = []
fft_cpu_rows = []
fft_gpu_rows = []

with open('../cuda-fft/result_fft.csv') as result_fft:
    reader = csv.reader(result_fft, quotechar='|')
    for row in reader:
        if (float(row[1]) >= 100):
            continue
        fft_n.append(row[0])
        fft_cpu_rows.append(float(row[1]))
        fft_gpu_rows.append(float(row[2]))
    
```

```python
fig, ax = plt.subplots(2, 2, layout='constrained', figsize=(20, 10))

ax[0, 0].plot(dft_n, dft_cpu_rows)
startx, endx = ax[0, 0].get_xlim()
starty, endy = ax[0, 0].get_ylim()
ax[0, 0].xaxis.set_ticks(np.arange(startx, endx, 20))
ax[0, 0].yaxis.set_ticks(np.arange(starty, endy, 100))
ax[0, 0].set_title('DFT CPU computation')
ax[0, 0].set_xlabel('N')
ax[0, 0].set_ylabel('Time used (ms)')

ax[0, 1].plot(dft_n, dft_gpu_rows)
startx, endx = ax[0, 1].get_xlim()
starty, endy = ax[0, 1].get_ylim()
ax[0, 1].xaxis.set_ticks(np.arange(startx, endx, 20))
ax[0, 1].yaxis.set_ticks(np.arange(starty, endy, 20))
ax[0, 1].set_title('DFT GPU computation')
ax[0, 1].set_xlabel('N')
ax[0, 1].set_ylabel('Time used (ms)')

ax[1, 0].plot(fft_n, fft_cpu_rows)
startx, endx = ax[1, 0].get_xlim()
starty, endy = ax[1, 0].get_ylim()
ax[1, 0].xaxis.set_ticks(np.arange(startx, endx, 200))
ax[1, 0].yaxis.set_ticks(np.arange(starty, endy, 5))
ax[1, 0].set_title('FFT CPU computation')
ax[1, 0].set_xlabel('N')
ax[1, 0].set_ylabel('Time used (ms)')

ax[1, 1].plot(fft_n, fft_gpu_rows)
startx, endx = ax[1, 1].get_xlim()
starty, endy = ax[1, 1].get_ylim()
ax[1, 1].xaxis.set_ticks(np.arange(startx, endx, 200))
ax[1, 1].yaxis.set_ticks(np.arange(starty, endy, 5))
ax[1, 1].set_title('FFT GPU computation')
ax[1, 1].set_xlabel('N')
ax[1, 1].set_ylabel('Time used (ms)')
```
